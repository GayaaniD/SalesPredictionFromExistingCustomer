```python
dataset=read_csv('DigitalAd_dataset.csv')
```

```python
print(dataset.shape)
```

```python
print(dataset.head(5))
```

```python
X=dataset.iloc[:,:-1].values
X
```

```python
Y=dataset.iloc[:,-1].values
Y
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,Y_train,Y_test = train_test_split(X,Y,test_size=0.25,random_state=0)
```

feature sacling

```python
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
X_train = sc.fit_transform(X_train)
X_test = sc.transform(X_test)
```

training

```python
from sklearn.linear_model import LogisticRegression
model = LogisticRegression(random_state=0)
model.fit(X_train,Y_train)
```

```python
Age=int(input("Enter new customer's age:"))
Sal=int(input("Enter new customer's salary:"))
NewCustomer = [[Age,Sal]]
result=model.predict(sc.transform(NewCustomer))
print(result)
if result == 1:
  print("Customer will buy")
else:
  print("Customer won't buy")
```

```python
y_predict = model.predict(X_test)
print(np.concatenate((y_predict.reshape(len(y_predict),1),Y_test.reshape(len(Y_test),1)),1))
```


Evaluating model - Confusion matrix

```python
from sklearn.metrics import confusion_matrix, accuracy_score
cm = confusion_matrix(Y_test,y_predict)

print("Confusion matrix:")
print(cm)

print("Accuracy of the model: {0}%".format(accuracy_score(Y_test,y_predict)*100))


```
